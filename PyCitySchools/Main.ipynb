```python
import pandas as pd
```

```python
schools_file = '../Resources/schools_complete.csv'
students_file = '../Resources/students_complete.csv'
```

```python
schools_df = pd.read_csv(schools_file)
schools_df.head()
```

```python
students_df = pd.read_csv(students_file)
students_df.head()
```

```python
total_schools = schools_df['School ID'].nunique()
total_schools
```

```python
students_df.nunique()
```

```python
total_students = students_df['Student ID'].nunique()
```

```python
total_students
```

```python
total_budget = schools_df['budget'].sum()
total_budget
```

```python
avg_math_score = students_df['math_score'].mean()
avg_reading_score = students_df['reading_score'].mean()
print(avg_math_score)
print(avg_reading_score)
```

```python
number_of_students_passing_math = students_df.loc[students_df['math_score']>=70]['Student ID'].nunique()
number_of_students_passing_math
```

```python
percent_students_passing_math = number_of_students_passing_math / total_students * 100
percent_students_passing_math
```

```python
number_of_students_passing_reading = students_df.loc[students_df['reading_score']>=70]['Student ID'].nunique()
number_of_students_passing_reading
```

```python
percent_students_passing_reading = number_of_students_passing_reading / total_students * 100
percent_students_passing_reading
```

```python
overall_passing_rate = (avg_math_score + avg_reading_score) / 2
overall_passing_rate
```

```python
district_summary_df = pd.DataFrame({'Total Schools': [total_schools],
                                  'Total Students': f"{total_students:,}",
                                   'Total Budget': f"${total_budget:,}",
                                   'Average Math Score': [avg_math_score],
                                   'Average Reading Score': [avg_reading_score],
                                   '% Passing Math': [percent_students_passing_math],
                                   '% Passing Reading': [percent_students_passing_reading],
                                   'Overall Passing Rate': [overall_passing_rate]})
district_summary_df.head()
```

```python
students_df_grouped_by_school = students_df.groupby('school_name').agg({'reading_score': ['mean', lambda x: (x >= 70).sum()/len(x)*100],
                                                                       'math_score': ['mean', lambda x: (x >= 70).sum()/len(x)*100]})
students_df_grouped_by_school.columns = ['Average Reading Score', '% Passing Reading', 'Average Math Score', '% Passing Math']
students_df_grouped_by_school['Overall_Passing_Rate'] = (students_df_grouped_by_school['% Passing Reading'] + students_df_grouped_by_school['% Passing Math']) / 2
students_df_grouped_by_school.head()
```

```python
merged_df = students_df_grouped_by_school.merge(schools_df, left_index=True, right_on='school_name').set_index('school_name')
merged_df.index.name = None
merged_df.head()
```

```python
#merged_df['Per Student Budget'] = (merged_df['budget'] / merged_df['size']).map('${:,.2f}'.format)
#merged_df['budget'] = merged_df['budget'].map('${:,.2f}'.format)
merged_df.rename(columns={'budget': 'Total School Budget', 'size': 'Total Students', 'type': 'School Type'}, inplace=True)
merged_df = merged_df[['School Type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Math Score', 'Average Reading Score',
                      '% Passing Math', '% Passing Reading', 'Overall_Passing_Rate']]
merged_df.head()
```
